```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("train.csv", index_col=0)
print(df.columns)
df
```

```python
df.isnull().sum()
```

```python
df1 = df["Sex"]
df1
```

```python
df_dummy = pd.get_dummies(df1,drop_first=True)
df["Sex_M"] = df_dummy #Male is one
df = df.drop(columns=["Sex"])
df
```

```python
df.isnull().sum()
```

```python
null_embarked = df[df["Embarked"].isnull()]
null_embarked
```

```python
from sklearn.preprocessing import LabelEncoder
encoded = df.apply(lambda col: LabelEncoder().fit_transform(df["Embarked"]) if col.dtype == "object" else col)
encoded.tail(65)
```

```python
df.isnull().sum()
```

```python
df = df.interpolate(method='polynomial', order=3)
df
```

```python
df.isnull().sum()
```

```python
df['Cabin'] = [str(x)[0] for x in df['Cabin']] #Converting cabins to numbers based on their fisrt alphabet
df
```

```python
le = LabelEncoder()
df['Cabin'] = le.fit_transform(df['Cabin'])
df['Embarked'] = le.fit_transform(df['Embarked'])
df
```

```python
df
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df2 = df.drop(columns={"Name","Ticket"})
df2
X = df2[["Pclass","Age","SibSp","Parch","Fare","Cabin","Embarked","Sex_M"]]
X
```

```python
y = df2["Survived"].values
y
```

```python
from sklearn.model_selection import train_test_split
#split dataset into train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=262)
X_train
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X_test
```

```python
import tensorflow as tf
import keras
tf.__version__
```

```python
### Initializing the ANN
ann = tf.keras.models.Sequential()
### Adding the input layer and the first hidden layer
ann.add(tf.keras.layers.Dense(units=20, activation='selu'))
### Adding the second hidden layer
ann.add(tf.keras.layers.Dense(units=20, activation='elu'))
### Adding the second hidden layer
ann.add(tf.keras.layers.Dense(units=5, activation='relu'))
### Adding the last hidden layer
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
### Compiling the ANN
ann.compile(optimizer = 'SGD', loss = 'binary_crossentropy', metrics = ['accuracy'])
### Training the ANN on the Training set
ann.fit(X_train, y_train, batch_size = 10, epochs = 400)
```

```python
from sklearn.metrics import average_precision_score
y_pred_train = np.round(ann.predict(X_train),0)
y_pred_test = np.round(ann.predict(X_test),0)
precision_test = average_precision_score(y_test, y_pred_test)
precision_train = average_precision_score(y_train, y_pred_train)
print(precision_train)
print(precision_test)
```

```python
from sklearn.metrics import f1_score
f1_train = f1_score(y_train, y_pred_train, average='macro')
f1_test = f1_score(y_test, y_pred_test, average='macro')
print(f1_train)
print(f1_test)
```

```python
### Initializing the ANN
ann_p = tf.keras.models.Sequential()
### Adding the input layer and the first hidden layer
ann_p.add(tf.keras.layers.Dense(units=80, activation='elu'))
### Adding the second hidden layer
ann_p.add(tf.keras.layers.Dense(units=16, activation='elu'))
### Adding the second hidden layer
ann_p.add(tf.keras.layers.Dense(units=8, activation='selu'))
### Adding the second hidden layer
ann_p.add(tf.keras.layers.Dense(units=32, activation='selu'))
### Adding the second hidden layer
ann_p.add(tf.keras.layers.Dense(units=8, activation='elu'))
### Adding the last hidden layer
ann_p.add(tf.keras.layers.Dense(units=16, activation='elu'))
### Adding the last hidden layer
ann_p.add(tf.keras.layers.Dense(units=8, activation='elu'))
### Adding the last hidden layer
ann_p.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
### Compiling the ANN
ann_p.compile(optimizer = 'RMSprop', loss = 'poisson', metrics = ['accuracy'])
### Training the ANN on the Training set
ann_p.fit(X_train, y_train, batch_size = 8, epochs = 150)
```

```python
from sklearn.metrics import average_precision_score
y_pred_train2 = np.round(ann_p.predict(X_train),0)
y_pred_test2 = np.round(ann_p.predict(X_test),0)
precision_test = average_precision_score(y_test, y_pred_test2)
precision_train = average_precision_score(y_train, y_pred_train2)
print(precision_train)
print(precision_test)
```

```python
from sklearn.metrics import f1_score
f1_train = f1_score(y_train, y_pred_train2, average='macro')
f1_test = f1_score(y_test, y_pred_test2, average='macro')
print(f1_train)
print(f1_test)
```
